```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt

# Dane
initial_price = 120000  # Pocztkowa cena mieszkania w PLN
growth_rate = 0.05  # Roczny wzrost cen mieszka (5%)
years = 5  # Okres oszczdzania w latach
months = years * 12  # Liczba miesicy
bank_rate = 0.12  # Nominalna roczna stopa procentowa banku (12%)
monthly_rate = bank_rate / 12  # Miesiczna stopa procentowa

# Przysza cena mieszkania
future_house_price = initial_price * (1 + growth_rate) ** years

# Miesicznej wpaty do banku
pmt = npf.pmt(rate=monthly_rate, nper=months, pv=0, fv=future_house_price)

# Dane do wykresu
monthly_deposit = pmt
time = np.arange(1, months + 1)

# Warto mieszkania w kadym miesicu (liniowy wzrost)
house_prices = initial_price * (1 + growth_rate) ** (time / 12)

# Warto lokaty na koniec kadego miesica
balance = npf.fv(rate=monthly_rate, nper=time, pmt=monthly_deposit, pv=0)

# Wykres
plt.figure(figsize=(10, 6))
plt.plot(time, house_prices, label='Cena mieszkania', color='red')
plt.plot(time, balance, label='Warto lokaty', color='green')
plt.title('Cena mieszkania i warto lokaty w czasie')
plt.xlabel('Miesice')
plt.ylabel('Warto (PLN)')
plt.legend()
plt.grid(True)
plt.show()

# Wyniki
print(f"Przewidywana cena mieszkania za 5 lat: {future_house_price:.2f} PLN")
print(f"Kwota miesicznej wpaty do banku: {monthly_deposit:.2f} PLN")
```
